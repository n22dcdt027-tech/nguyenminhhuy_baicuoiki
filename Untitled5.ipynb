```python
# ==============================================
# HIỂN THỊ ẢNH KẾT QUẢ + BIỂU ĐỒ TRONG COLAB
# (Multi-image inference with LLaVA-1.5-7B, 4-bit)
# ==============================================
import os, subprocess, shlex, time, textwrap
from pathlib import Path
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
from google.colab import files

# ---------- A) CÀI MÔI TRƯỜNG (nếu chưa có) ----------
def ensure_setup():
    if not Path("/content/LLaVA").exists():
        print(">> Cài đặt LLaVA và phụ thuộc ...")
        os.system("pip -q uninstall -y torch torchvision torchaudio")
        os.system("pip -q install --index-url https://download.pytorch.org/whl/cu118 torch==2.3.1 torchvision==0.18.1 torchaudio==2.3.1")
        os.system("git clone https://github.com/haotian-liu/LLaVA.git /content/LLaVA")
        os.system("pip -q install -e /content/LLaVA --no-deps")
        os.system('pip -q install --no-deps "transformers==4.45.2" "tokenizers==0.19.1" "accelerate==0.33.0" '
                  '"sentencepiece==0.2.0" "safetensors>=0.4.4" "bitsandbytes==0.43.1" "peft==0.12.0" '
                  '"einops==0.6.1" "einops-exts==0.0.4" "timm==0.9.16" "shortuuid==1.0.13" '
                  '"markdown2==2.4.12" "httpx==0.24.1" "gradio==4.44.0" "opencv-python-headless==4.10.0.84"')
        print(">> Setup xong.")

ensure_setup()

# ---------- B) HÀM CHẠY LLaVA & VẼ ẢNH KẾT QUẢ ----------
MODEL = "liuhaotian/llava-v1.5-7b"
OUT_DIR = Path("/content/llava_results"); OUT_DIR.mkdir(parents=True, exist_ok=True)

def run_llava(image_path, prompt):
    cmd = f"""python -m llava.serve.cli \
      --model-path {MODEL} \
      --image-file "{image_path}" \
      --load-4bit \
      --prompt "{prompt}" """
    t0 = time.time()
    p = subprocess.run(shlex.split(cmd), capture_output=True, text=True, cwd="/content/LLaVA")
    sec = time.time() - t0
    stdout = (p.stdout or "").strip()
    # lấy dòng cuối làm câu trả lời
    lines = [ln.strip() for ln in stdout.splitlines() if ln.strip()]
    answer = lines[-1] if lines else ""
    return answer, sec, stdout

def draw_card(src_img_path, question, answer, save_path, max_w=1400):
    img = Image.open(src_img_path).convert("RGB")
    # scale ảnh về chiều rộng hợp lý
    w = min(max_w//2, 800)
    ratio = w / img.width
    img = img.resize((w, int(img.height * ratio)))
    pad = 24
    text_box_w = max_w - (img.width + 3*pad)

    # fonts
    try:
        font = ImageFont.truetype("DejaVuSans.ttf", 24)
        font_b = ImageFont.truetype("DejaVuSans-Bold.ttf", 26)
    except:
        font = ImageFont.load_default()
        font_b = ImageFont.load_default()

    # wrap text
    def wrap(s, width=46):
        out = []
        for line in s.splitlines():
            out += textwrap.wrap(line, width=width) or [""]
        return "\n".join(out)

    q_txt = wrap(f"Q: {question}", 46)
    a_txt = wrap(f"A: {answer}", 46)

    # estimate text height
    tmp = Image.new("RGB", (10,10), "white")
    dtmp = ImageDraw.Draw(tmp)
    hq = dtmp.multiline_textbbox((0,0), q_txt, font=font_b, spacing=6)[3]
    ha = dtmp.multiline_textbbox((0,0), a_txt, font=font, spacing=6)[3]
    text_h = hq + ha + 3*pad

    canvas_h = max(img.height + 2*pad, text_h + 2*pad)
    canvas = Image.new("RGB", (max_w, canvas_h), "white")
    draw = ImageDraw.Draw(canvas)

    # paste image
    canvas.paste(img, (pad, (canvas_h - img.height)//2))

    # draw text box
    x0 = img.width + 2*pad
    y0 = pad
    draw.rectangle([x0-10, y0-10, x0 + text_box_w + 10, y0 + text_h - pad], outline="black", width=1)
    draw.text((x0, y0), q_txt, font=font_b, fill="black", spacing=6)
    draw.text((x0, y0 + hq + pad//2), a_txt, font=font, fill="black", spacing=6)

    canvas.save(save_path)
    return save_path

# ---------- C) CHUẨN BỊ ẢNH MẪU + ẢNH NGƯỜI DÙNG (TÙY CHỌN) ----------
os.makedirs("/content/images", exist_ok=True)
# ảnh mẫu (tự tải)
os.system('wget -q https://llava-vl.github.io/static/images/monalisa.jpg -O /content/images/monalisa.jpg')
os.system('wget -q https://llava-vl.github.io/static/images/view.jpg -O /content/images/view.jpg')

# cho phép upload thêm ảnh (không bắt buộc)
print("Nếu muốn thêm ảnh của bạn, chọn file ở hộp thoại (có thể bỏ qua).")
try:
    up = files.upload()
    for k in up.keys():
        Path(f"/content/images/{k}").write_bytes(up[k])
except Exception:
    pass

# danh sách ảnh & câu hỏi
inputs = [
    ("/content/images/monalisa.jpg", "Đây là bức tranh gì? Hãy nêu 1 đặc điểm nổi bật."),
    ("/content/images/view.jpg", "Mô tả cảnh trong ảnh và đếm có bao nhiêu chiếc thuyền?"),
]
# thêm ảnh người dùng nếu có
for p in sorted(Path("/content/images").glob("*")):
    if p.name not in {"monalisa.jpg","view.jpg"} and p.suffix.lower() in {".jpg",".jpeg",".png"}:
        inputs.append((str(p), "Hãy mô tả đối tượng chính và màu sắc nổi bật trong ảnh."))

# ---------- D) CHẠY & HIỂN THỊ KẾT QUẢ ----------
records = []  # (name, prompt, answer, sec, word_count, card_path)
display(Markdown("## ✅ Kết quả từng ảnh"))
for img_path, prompt in inputs:
    ans, sec, raw = run_llava(img_path, prompt)
    name = Path(img_path).stem
    card_path = OUT_DIR / f"result_{name}.png"
    draw_card(img_path, prompt, ans, str(card_path))

    # hiển thị ngay trong kết quả
    display(Markdown(f"**Ảnh:** `{name}` — **Thời gian:** {sec:.2f}s — **Số từ:** {len(ans.split())}"))
    display(Image.open(card_path))

    records.append((name, prompt, ans, sec, len(ans.split()), str(card_path)))

# ---------- E) VẼ BIỂU ĐỒ (độ dài câu trả lời + thời gian) ----------
names = [r[0] for r in records]
lengths = [r[4] for r in records]
secs = [r[3] for r in records]

plt.figure(figsize=(8,4))
plt.bar(names, lengths)  # không đặt màu theo yêu cầu
plt.title("Độ dài câu trả lời (số từ)")
plt.xlabel("Ảnh"); plt.ylabel("Số từ")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
plt.bar(names, secs)  # không đặt màu theo yêu cầu
plt.title("Thời gian suy luận (giây)")
plt.xlabel("Ảnh"); plt.ylabel("Giây")
plt.tight_layout()
plt.show()

print("\nCác file ảnh kết quả đã tạo trong:", OUT_DIR)
for p in sorted(OUT_DIR.glob("*.png")):
    print(" -", p)
```
